```python
import tensorflow as tf
import numpy as np
import pandas as pd
from keras.layers import Dense
from keras.models import Sequential
from keras.utils import to_categorical
import matplotlib as plt
```

```python
df=pd.read_csv("D:\VERZEO (AI ML)\AI ML\HEART DISEASE\heart.csv")   #to load the dataset of ML model
df.head()
```

```python
df
```

```python
y=df["target"]     #storing the target column in the Y variable
```

```python
target=to_categorical(y)
```

```python
X=df.drop("target",axis=1)   #now X contains all the column except "target". "axis=1" indicates python to search in columns
```

```python
X   #now X just contains the features 
```

# Specify the Architecture

```python
model=Sequential()
model.add(Dense(200,activation="relu", input_shape=(13,)))   #13 signifies your feature columns
model.add(Dense(200,activation="relu")) 
model.add(Dense(300,activation="relu"))
model.add(Dense(2, activation="softmax"))
```

```python
model.compile(optimizer="adam", loss="categorical_crossentropy",metrics=["accuracy"])   
#adam initiates Gradient Descent "mean square" to ananlyse loss
```

# Fit the model

```python
model.fit(X,target, epochs=50) #always pass Features,Targets therefore X,y "epochs" to know how much time to run
```
